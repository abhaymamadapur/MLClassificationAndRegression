```python
trainDf = df.copy()
```

```python
targetData = trainDf['Target']
inputData = trainDf.drop(['Target'],axis=1)
```

### Splitting of data set 

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(inputData,targetData,test_size =0.3,random_state=0)
```

```python
from sklearn import preprocessing

def normalizeDataset(dataset):
    x = dataset.values #returns a numpy array
    std_scaler = preprocessing.MinMaxScaler()
    x_scaled = std_scaler.fit_transform(x)
    return pd.DataFrame(x_scaled)
```

```python
X_train = normalizeDataset(X_train)
X_test = normalizeDataset(X_test)
```

###  Using Linear Regression 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error
lr_model = LinearRegression()
lr_model.fit(X_train,y_train)
```

```python
predicted_val = lr_model.predict(X_test)
predicted_val
```

```python
lr_score = lr_model.score(X_test,y_test)
lr_score
```

```python
lr_sqrt = np.sqrt(mean_squared_error(y_test,predicted_val))
lr_sqrt
```

```python
lr_r2 = r2_score(y_test,predicted_val)
lr_r2
```

### Using SVM Regression

```python
from sklearn.svm import SVR
from sklearn.metrics import r2_score,mean_squared_error
svr_model = SVR(gamma='auto',kernel='poly')
svr_model.fit(X_train,y_train)
```

```python
svr_prediction_val = svr_model.predict(X_test)
svr_prediction_val
```

```python
svr_score = svr_model.score(X_test,y_test)
svr_score
```

```python
np.sqrt(mean_squared_error(y_test,svr_prediction_val))
```

```python
r2_score(y_test,svr_prediction_val)
```

### Using Random Forest 

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=184,
                                  min_samples_split=2,
                                  min_samples_leaf=1,
                                  max_features='sqrt',
                                  max_depth=20,
                                  bootstrap=False)
regressor.fit(X_train, y_train)
```

```python
regressor_pred = regressor.predict(X_test)
regressor_pred
```

```python
reg_score = regressor.score(X_test,y_test)
reg_score
```

```python
np.sqrt(mean_squared_error(y_test,regressor_pred))
```

```python
r2_score(y_test,regressor_pred)
```
